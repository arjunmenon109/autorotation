```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constants Below
d = 1.293 # density of air 
c = 1.5  # Coeficient Of Drag
pi = 3.14159 # Pie
# Variables 
wh = 0.5
wm = 0.5 # total weight of mass
w0 = 0.05 # initial weight of rotors
r = 0.1 # rotor radius
r0 = 0.1 # init radius of rotor
# Used in equations
v = []
v2 = []
x1 = []
x2 = []
x3=[]
```

Corelation between increase in mass and velocity at steady state

```python
while wh<=5:
  i = math.sqrt((2*wh)/(d*pi*(r*r)*c))
  v.append(i)
  x1.append(wh)
  wh = wh+0.1
```

```python
fig, ax = plt.subplots()
plt.xlabel("Mass")
plt.ylabel("Velocity at Steady State")
ax.plot(x1, v)
plt.show()
```

We have to account for increase in weight as the radius also increases. This is done by using the following relation:
image.png



```python
while r<=1:
  wr=w0*((r/r0)*(r/r0)*(r/r0))
  w=wm+wr
  i = math.sqrt((2*w)/(d*pi*(r*r)*c))
  v2.append(i)
  x3.append(w)
  x2.append(r)
  r = r+0.01
```

```python
fig, ay = plt.subplots()
ay.plot(x2, v2)
plt.xlabel("Radius")
plt.ylabel("Velocity at Steady State")
plt.show()

fig, az = plt.subplots()
az.plot(x2, x3)
plt.xlabel("Radius")
plt.ylabel("Exponential Increase in Mass")
plt.show()
```
